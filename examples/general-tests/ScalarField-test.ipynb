```julia
using LowLevelFEM
```

```julia
structured_rect_mesh()

mat = Material("body")
prob = Problem([mat], type=:ScalarField, dim=2, bandwidth=:RCMK)
```

```julia
s = ScalarField(prob, "body", (x, y, z)->x)
```

```julia
s2 = scalarField(prob, "body", (x, y, z)->x)
s2 = nodesToElements(s2)
```

```julia
showElementResults(s, visible=true)
showElementResults(s2)
```

```julia
v = VectorField(prob, "body", [(x, y, z)->x, (x, y, z)->y, (x, y, z)->z])
```

```julia
showElementResults(v)
```
