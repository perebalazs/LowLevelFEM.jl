```julia
using LowLevelFEM
```

```julia
gmsh.initialize()
gmsh.open("cube1.geo")

mat = material("cube")
prob = Problem([mat])
```

```julia
s = ScalarField(prob, "cube", (x, y, z, t) -> t^3, steps=10)
```

```julia
dsdt = ∂t(s)
```

```julia
s2 = ∫(dsdt)
```

```julia
s = elementsToNodes(s)
```

```julia
dsdt = ∂t(s)
```

```julia
s2 = ∫(dsdt)
```

```julia
s2 = ∫(∂t(s))
```

```julia
using LinearAlgebra
integrate(prob, "cube", (s - s2) * (s - s2), step=10)
```
