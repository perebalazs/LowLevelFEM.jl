```julia
using LowLevelFEM
```

```julia
gmsh.initialize()
gmsh.open("cube1.geo")

mat = material("cube")
prob = Problem([mat])
```

```julia
s = scalarField(prob, "cube", (x, y, z) -> x^3);
ds = ∂x(elementsToNodes(∂x(s)))
```

```julia
S = ScalarField(prob, "cube", (x, y, z) -> x^3);
dS = ∂x(∂x(S))
```

```julia
v = VectorField([s, 2s, 3s])
```

```julia
V = VectorField([S, 2S, 3S])
```
