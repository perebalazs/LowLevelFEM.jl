```julia
using LowLevelFEM
```

```julia
gmsh.initialize()
gmsh.open("cube1.geo")

mat = material("cube")
prob = Problem([mat])
```

```julia
s = scalarField(prob, "cube", (x, y, z) -> x^2);
ds = ∂x(∂x(s))
```

```julia
vs = [s, s, s]
typeof(vs)
```

```julia
v = VectorField(vs)
```

```julia
@showstruct v
```

```julia
s1 = nodesToElements(s)
v1 = [s1, 2s1, 3s1];
elems = v1[1].numElem
ε = Vector{Matrix{Float64}}(undef, length(elems))
for i in 1:length(elems)
    # comps1[j].a[i] is a (numNodes × nsteps) matrix
    blocks = [v1[j].A[i] for j in 1:3]
    ε[i] = vcat(blocks...)   # 3*numNodes × nsteps
end
ε
```

```julia
showElementResults(ds)
openPostProcessor()
```

```julia
gmsh.finalize()
```
