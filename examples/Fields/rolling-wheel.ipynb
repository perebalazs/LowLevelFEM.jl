```julia
using LowLevelFEM
```

```julia
gmsh.initialize()
gmsh.open("wheel.geo")

mat = material("wheel")
prob = Problem([mat])
```

```julia
st = 1000
ω = 10
r = 5
fc(x, y, z, t) = cos(-ω * t)
fs(x, y, z, t) = sin(-ω * t)

S = ScalarField(prob, "wheel", fs, steps=st, tmax=1)
C = ScalarField(prob, "wheel", fc, steps=st, tmax=1)
I = ScalarField(prob, "wheel", 1, steps=st, tmax=1)
O = ScalarField(prob, "wheel", 0, steps=st, tmax=1)

R = TensorField(
    [C O -S
        O I O
        S O C])

fx(x, y, z, t) = ω * r * t

X = ScalarField(prob, "wheel", fx, steps=st, tmax=1)
O = ScalarField(prob, "wheel", 0, steps=st, tmax=1)
H = ScalarField(prob, "wheel", r, steps=st, tmax=1)

C = VectorField([X, O, O])
h = VectorField([O, O, H]);
```

```julia
x = ScalarField(prob, "wheel", (x, y, z, t) -> x, steps=st, tmax=1)
y = ScalarField(prob, "wheel", (x, y, z, t) -> y, steps=st, tmax=1)
z = ScalarField(prob, "wheel", (x, y, z, t) -> z, steps=st, tmax=1)

r0 = VectorField([x, y, z]);
```

```julia
rt = R * (r0 - h) + h + C

showDeformationResults(rt)
```

```julia
gmsh.finalize()
```
