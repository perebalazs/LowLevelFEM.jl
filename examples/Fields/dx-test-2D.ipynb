```julia
using LowLevelFEM
```

```julia
gmsh.initialize()
gmsh.open("square.geo")
```

```julia
mat = material("surf")
prob = Problem([mat], type=:PlaneStress)
```

```julia
s = scalarField(prob, "surf", (x, y, z) -> x)

ds = ∂x(s)
```

```julia
s = ScalarField(prob, "surf", (x, y, z) -> x)

ds = ∂x(s)
```

```julia
gmsh.finalize()
```
