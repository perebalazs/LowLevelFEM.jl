```julia
using LowLevelFEM
```

```julia
structured_rect_mesh()

mat = Material("body", E=2e5, ν=0.3)
prob = Problem([mat], type=:PlaneStrain)  # :Solid, :PlaneStress, :AxiSymmetric, :HeatConduction, ...

bc = displacementConstraint("left", ux=0, uy=0)
force = load("right", fy=-1)

u = solveDisplacement(prob, load=[force], support=[bc])
S = solveStress(u)

showDoFResults(u)
showDoFResults(u, :ux)
showStressResults(S)
showStressResults(S, :sxy, name="Shear stress")

openPostProcessor()
```

```julia
K = stiffnessMatrix(prob)
f = loadVector(prob, [force])
u2 = solveDisplacement(K, f, support=[bc])

E = mat.E
ν = mat.ν

u2 = expandTo3D(u2)

A = (u2 ∘ ∇ + ∇ ∘ u2) / 2
I = TensorField(prob, "body", [1 0 0; 0 1 0; 0 0 1])
S2 = E / (1 + ν) * (A + ν / (1 - 2ν) * trace(A) * I)

showDoFResults(u2)
showDoFResults(u2, :ux)
showStressResults(S2)
showStressResults(S2, :sxy, name="Shear stress")

openPostProcessor()
```
