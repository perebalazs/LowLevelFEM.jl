```julia
using LowLevelFEM
```

```julia
openGeometry("truss2D.geo")

#openPreProcessor(openGL=true)
```

```julia
mat = Material("rod")
Pu = Problem([mat], type=:VectorField, dim=2, field=:u, rhs_field=:f)
```

```julia
A = mat.A
E = mat.E
```

```julia
K = ∫(AxialGrad(Pu) ⋅ (A * E) ⋅ AxialGrad(Pu))
K[:, :]
```

```julia
fR = ∫(Pu ⋅ [0, -1], Γ="3")
```

```julia
supp1 = BoundaryCondition("1", ux=0, uy=0)
supp2 = BoundaryCondition("2", ux=0, uy=0)
```

```julia
u = solveField(K, fR, support=[supp1, supp2])
u.a
```

```julia
t = tangentVector(Pu, "rod")
```

```julia
g = grad(expandTo3D(u))
```

```julia
ε = t * (g * t)
```

```julia
N = E * A * ε
```

```julia
N.A
```

```julia
showDoFResults(u, name="u")
showElementResults(N, name="N")
```
