```julia
using LowLevelFEM
```

```julia
openGeometry("beam-1.geo")
#openPreProcessor()

mat = Material("beam")
Pu = Problem([mat], type=:VectorField, dim=2, field=:u, rhs_field=:f)
Pφ = Problem([mat], type=:ScalarField, dim=2, field=:φ, rhs_field=:m)
```

```julia
E = mat.E
G = mat.μ
I = 1
κ = 2 / 3
A = 1
```

```julia
t = tangentVector(Pu, "beam")
tx, ty, tz = t[1], t[2], t[3]

nx = -ty
ny = tx
nz = 0
```

```julia
a = [tx * tx tx * ty ty * tx ty * ty]   # axial projector
s = [nx * tx nx * ty ny * tx ny * ty]   # shear projector
```

```julia
Cuu = E * A * (a' * a) + κ * G * A * (s' * s)

Cuφ = -κ * G * A * s'   # 4×1
Cφu = -κ * G * A * s    # 1×4
```

```julia
Kuu = ∫(Grad(Pu) ⋅ Cuu ⋅ Grad(Pu); Γ="beam")

Kuφ = ∫(Grad(Pu) ⋅ Cuφ ⋅ Id(Pφ); Γ="beam")
Kφu = ∫(Id(Pφ) ⋅ Cφu ⋅ Grad(Pu); Γ="beam")

Kφφ = ∫(Id(Pφ) ⋅ (κ * G * A) ⋅ Id(Pφ); Γ="beam") +
      ∫(AxialGrad(Pφ) ⋅ (E * I) ⋅ AxialGrad(Pφ); Γ="beam")
```

```julia
K = SystemMatrix([Kuu Kuφ; Kφu Kφφ])
```

```julia
supp_u = BoundaryCondition("A", problem=Pu, ux=0, uy=0)
supp_φ = BoundaryCondition("A", problem=Pφ, φ=0)
```

```julia
fu = ∫(Pu ⋅ [1, 0], Γ="C")
fφ = ∫(Pφ ⋅ 0, Γ="B")
fu2 = ∫(Pu ⋅ [0.0, 0], Γ="beam")
```

```julia
F = SystemVector([fu + fu2, fφ])
```

```julia
#u, v, φ = solveField(K, F, support=[supp_u, supp_v, supp_φ, supp_u0, supp_v0, supp_φ0])
u, φ = solveField(K, F, support=[supp_u, supp_φ])
```

```julia
showDoFResults(fu, name="force")
u0 = showDoFResults(u, name="u", factor=10, visible=true)
φ0 = showDoFResults(φ, name="φ")
```

```julia
n = VectorField([nx, ny, 0nx])
```

```julia
g = grad(expandTo3D(u))

ε = t ⋅ (g * t)
γ = n ⋅ (g * t) - nodesToElements(φ)

κb = grad(φ) ⋅ t

N = E * A * ε
T = κ * G * A * γ
Mh = E * I * κb
```

```julia
N0 = showElementResults(N, name="N")
T0 = showElementResults(T, name="T")
Mh0 = showElementResults(Mh, name="Mh")
```

```julia
plotOnPath("beam", N0)
plotOnPath("beam", T0)
plotOnPath("beam", Mh0)
```
