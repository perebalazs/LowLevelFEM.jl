```julia
using LowLevelFEM
```

```julia
gmsh.initialize()
gmsh.open("cube1.geo")

mat = material("cube")
prob = Problem([mat])
```

```julia
R00 = tensorField(prob, "cube", [0 0 0; 0 0 0; 0 0 0])
RR = zeros(size(R00.a, 1), 100)

for i in 1:100
    φ = i / 100 * π
    f1(x, y, z) = cos(φ)
    f2(x, y, z) = -sin(φ)
    f3(x, y, z) = sin(φ)
    R0 = [f1 f2 0; f3 f1 0; 0 0 1]
    R1 = tensorField(prob, "cube", R0)
    RR[:, i] = R1.a
end

t = [i for i in 1:100]
R = TensorField([], RR, t, [], 100, :tensor, R00.model)
```

```julia
r = nodePositionVector(prob)
r0 = zeros(size(r.a, 1), 100)
for i in 1:100
    r0[:, i] = r.a
end
r = VectorField([], r0, t, [], 100, :v3D, r.model)
r = nodesToElements(r)
```

```julia
R = nodesToElements(R)
```

```julia
showDeformationResults(R * r)
```
