```julia
using LowLevelFEM
```

```julia
gmsh.initialize()
gmsh.open("BC-tests.geo")

mat = Material("volu")
prob = Problem([mat], type=:General, dim=3, fdim=1)
```

```julia
nodes, _, _ = gmsh.model.mesh.getNodes()
length(nodes)
```

```julia
BC = BoundaryCondition("left", ux=1, uy=0, uz=0)
```

```julia
bc = constrainedDoFs(prob, [BC])
length(bc)
bc
```

```julia
all = allDoFs(prob)
length(all)
all
```

```julia
free = freeDoFs(prob, [BC])
length(free)
free
```

```julia
a = scalarField(prob, "volu", 1)
showDoFResults(a, name="volu")
```

```julia
b = scalarField(prob, "left", 1)
showDoFResults(b, name="left")
```

```julia
c = applyBoundaryConditions(prob, [BC])
showDoFResults(c, name="apply BC")
```

```julia
openPostProcessor()
```

```julia
gmsh.finalize()
```
