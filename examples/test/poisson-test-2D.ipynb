```julia
using LowLevelFEM
```

```julia
gmsh.initialize()
gmsh.open("rod.geo")

mat = material("body")
prob = Problem([mat], type=:Poisson2D)
```

```julia
s = ScalarField(prob, "body", 1.0);
```

```julia
@time M1 = LowLevelFEM.stiffnessMatrixPoisson(prob);
```

```julia
@time M2 = LowLevelFEM.stiffnessMatrixPoisson(prob, coefficient=s);
```

```julia
using SparseArrays, LinearAlgebra
norm(M1.A - M2.A)
```

```julia
gmsh.finalize()
```
