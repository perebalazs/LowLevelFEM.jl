```julia
using LowLevelFEM
```

```julia
gmsh.initialize()
gmsh.open("cube1.geo")

mat = material("cube")
prob = Problem([mat], type=:Poisson)
```

```julia
s = ScalarField(prob, "cube", 1.0);
```

```julia
a = Dict(zip(s.numElem, s.A));

#get(a, 1, 1.0)
```

```julia
M1 = LowLevelFEM.systemMatrixPoisson(prob, 1)
```

```julia
M2 = LowLevelFEM.systemMatrixPoisson2(prob, s)
```

```julia
using SparseArrays, LinearAlgebra
norm(M1.A - M2.A)
```

```julia
gmsh.finalize()
```
