```julia
using LowLevelFEM
```

```julia
gmsh.initialize()
gmsh.open("cube1.geo")

mat = material("cube")
prob = Problem([mat], type=:Poisson)
```

```julia
s = ScalarField(prob, "cube", 1.0);
```

```julia
@time M1 = LowLevelFEM.stiffnessMatrixPoisson(prob, coefficient=1);
```

```julia
@time M2 = LowLevelFEM.stiffnessMatrixPoisson(prob, coefficient=s);
```

```julia
using SparseArrays, LinearAlgebra
norm(M1.A - M2.A)
```

```julia
gmsh.finalize()
```
