```julia
import LowLevelFEM as FEM
using LowLevelFEM
```

```julia
gmsh.initialize()
```

```julia
gmsh.open("valve.geo")
mat = FEM.material("body", E=2.e5, ν=0.3)
problem = FEM.Problem([mat], type=:AxiSymmetric)
```

```julia
supp = FEM.displacementConstraint("supp", uy=0)
```

```julia
K = FEM.stiffnessMatrix(problem)
```

```julia
M = FEM.massMatrix(problem)
```

```julia
Tₘᵢₙ = FEM.smallestPeriodTime(K, M)
ωₘₐₓ = 2π / Tₘᵢₙ
```

```julia
C = FEM.dampingMatrix(K, M, ωₘₐₓ, β=0)
```

```julia
dof, dof = size(K)
```

```julia
f = zeros(dof)
```

```julia
FEM.applyBoundaryConditions!(problem, K, M, C, f, [supp])
```

```julia
u0 = zeros(dof)
v0 = zeros(dof)
FEM.initialVelocity!(problem, "body", v0, vy=1000)
FEM.initialVelocity!(problem, "supp", v0, vy=0)
```

```julia
E = problem.material[1].E
ρ = problem.material[1].ρ
c = √(E / ρ)
```

```julia
u, v, t = FEM.CDM(K, M, C, f, u0, v0, 120 / c, Tₘᵢₙ / π)
```

```julia
S = FEM.solveStress(problem, u)
```

```julia
vvec = FEM.showDoFResults(problem, v, t=t, :vvec, name="v(t)", visible=false)
```

```julia
seqv = FEM.showStressResults(problem, S, t=t, :seqv, name="σeqv", visible=true)
```

```julia
gmsh.fltk.run()
```

```julia
gmsh.finalize()
```
