```julia
using SparseArrays

function stokesMatrix(K::LowLevelFEM.SystemMatrix, G::LowLevelFEM.SystemMatrix, prob_v::Problem, prob_p::Problem)

    nu = prob_v.dim * prob_v.non
    np = prob_p.non

    Zpp = spzeros(np, np)

    A = [
        K.A G.A;
        -G.A' Zpp
    ]

    return A
end
```

```julia
function stokesRHS(f_u::VectorField, prob_v::Problem, prob_p::Problem)
    fu = f_u.a[:, 1]
    fp = zeros(prob_p.non)
    return vcat(fu, fp)
end

```

```julia
K = symmetricGradientMatrix(prob_v)
G = gradMatrix(prob_v, prob_p)

A = stokesMatrix(K, G, prob_v, prob_p)
b = stokesRHS(f_u, prob_v, prob_p)

nu = prob_v.dim * prob_v.non
A[nu+1, :] .= 0
A[:, nu+1] .= 0
A[nu+1, nu+1] = 1
b[nu+1] = 0

x = A \ b

u_sol = x[1:nu]
p_sol = x[nu+1:end]
```
