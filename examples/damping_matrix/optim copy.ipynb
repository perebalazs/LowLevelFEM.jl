```julia
using Optimization, OptimizationOptimJL
#rosenbrock(x, p) = (p[1] - x[1])^2 + p[2] * (x[2] - x[1]^2)^2
#ff(x, p) = (x[1] - p[1])^2 + (x[2] - p[2])^2
function ff(x, p)
    y = 0
    for z in 0:0.1:3
        y += (sinh(z) - (x[1] + x[2] * z + x[3] * z^2 + x[4] * z^3 + x[5] * z^4))^2
    end
    return y
end
x0 = zeros(5)
p = 0
f = OptimizationFunction(ff)
prob = Optimization.OptimizationProblem(f, x0, p, lb=[-100.0, -100.0, -100.0, -100.0, -100.0], ub=[100.0, 100.0, 100.0, 100.0, 100.0])
sol = solve(prob, Optim.ParticleSwarm(lower=prob.lb, upper=prob.ub, n_particles=100))
```

```julia
x = 0:0.1:3
plot(x, sinh.(x))
fs(x) = sol.u[1] + sol.u[2] * x + sol.u[3] * x^2 + sol.u[4] * x^3 + sol.u[5] * x^4



plot!(x, fs.(x))
```
